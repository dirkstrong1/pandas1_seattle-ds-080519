Here's a row that our engineer accidentally left out of the .csv file, expressed as a Python dictionary:

```python
extra_rows = {'age': [40, 30], 'sex': [1, 0], 'cp': [0, 0], 'trestbps': [120, 130], 'chol': [240, 200],
             'fbs': [0, 0], 'restecg': [1, 0], 'thalach': [120, 122], 'exang': [0, 1],
              'oldpeak': [0.1, 1.0], 'slope': [1, 1], 'ca': [0, 1], 'thal': [2, 3], 'target': [0, 0]}
extra_rows
```

How can we add this to the bottom of our dataset?

```python
two_rows = pd.DataFrame(extra_rows)

heart_augmented = pd.concat([heart, two_rows])


heart_augmented_fixed = heart_augmented.reset_index(drop = True)

heart_augmented_fixed.tail()
```

How many different values does have slope have? What about sex? And target?

```python
heart_augmented_fixed['sex'].value_counts()
```

Let's add a new column to our dataset called "test". Set all of its values to 0.

```python
heart_augmented_fixed['test'] = 0


heart_augmented_fixed
```

## Filtering

We can use filtering techniques to see only certain rows of our data. If we wanted to see only the rows for patients 70 years of age or older, we can simply type:

```python
heart_augmented_fixed[heart_augmented_fixed['age'] >= 70]
```

Use '&' for "and" and '|' for "or".

```python
heart_augmented_fixed[(heart_augmented_fixed['age'] >= 70) | (heart['trestbps'] > 170)]
```

## .loc() and .iloc()

We can use .loc to get, say, the first ten values of the age and trestbps columns:

```python
heart.loc[:10, ['age', 'trestbps']]
```

.iloc() is used for selecting locations in the DataFrame **by number**:

```python
# How would we get the same slice as just above by using .iloc() instead of .loc()?
```

```python
heart_augmented_fixed.iloc[:10, [0, 3]]
```
